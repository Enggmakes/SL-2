```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
class AF():
    def linear(x):
        return x
    def sigmoid(x):
        return 1/(1+np.exp(-x))
    def tanh(x):
        return np.tanh(x)
    def RELU(x):
        x1=[]
        for i in x:
            if i<0:
                x1.append(0)
            else:
                x1.append(i)
        return x1
    def softmax(x):
        return np.exp(x)/np.sum(np.exp(x),axis=0)
```

```python
while True:
    n=int(input("Enter \n0. to cancel\n1. Linear\n2. sigmoid\n3. tanh\n4. RELU\n5. Softmax"))
    if n==0:
        break
    if n==1:
        x=np.linspace(-10,10)
        plt.plot(x,AF.linear(x))
        plt.title("LINEAR")
        plt.show()
    elif n==2:
        x=np.linspace(-10,10)
        plt.plot(x,AF.sigmoid(x))
        plt.title("SIGMOID")
        plt.show()
    elif n==3:
        x=np.linspace(-10,10)
        plt.plot(x,AF.tanh(x))
        plt.title("TANH")
        plt.show()
    elif n==4:
        x=np.linspace(-10,10)
        plt.plot(x,AF.RELU(x))
        plt.title("RELU")
        plt.show()
    elif n==5:
        x=np.linspace(-10,10)
        plt.plot(x,AF.softmax(x))
        plt.title("SOFTMAX")
        plt.show()      
    
```
